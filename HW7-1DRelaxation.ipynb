```python
##oh you know
import numpy as np
import matplotlib.pyplot as plt
import math

plt.style.use('ggplot')
%matplotlib inline
```

```python
##Initial conditions
s = 100
x = np.linspace(0, s)
V = np.zeros_like(x) ##Look at this awful guess 
a = 1

##Boundries 
V[0] = 0
V[-1] = s

##Error
Err = 10**(-4)
E_t = 100 

##Interation check
k = 0
max_it = 100

E_lst = np.zeros(max_it)
lst = np.zeros_like(E_lst)
##Do many times to decrease error and have a max iter check
while k<max_it or Err>E_t:
##Iterate through each point in the mesh
    E_t = 0 
    for i in range(1,s-1):
        try:
            V[i] = 1/2*(V[i+a]+V[i-a])
            ##Bc it's linear, V[i]=i ideally
            E_t += abs(V[i]-i)/i
        except IndexError:
            break 
    plt.plot(x,V)
    E_lst[k] = E_t
    k+=1
    lst[k-1]=k
```

```python
##Plotting Final
plt.figure(figsize=(8,8))
plt.plot(x,V)
plt.xlabel('x')
plt.ylabel('V(x)')
plt.xlim([0,s])
plt.ylim([0,V.max()]) 
plt.title("100 iterations")
```

```python
##Plotting Error
plt.figure(figsize=(8,8))
plt.plot(E_lst)
plt.xlabel('Iterations')
plt.ylabel('Error')
plt.xlim([0,100])
plt.ylim([25,50]) 
plt.title("Error after 100 interations")
#print(E_lst)
```
